## Merging both datasets

```python
import pandas as pd

df_fifa_features_all = pd.read_csv('df_fifa_features_all.csv')
df_stanford_real_features = pd.read_csv('df_stanford_real_features.csv')
df_total_features = pd.concat([df_fifa_features_all,df_stanford_real_features], axis=1,sort=False)
```

```python
df_total_features=df_fifa_features_all.merge(df_stanford_real_features, left_on='id', right_on='id',how='left')
```

```python
import pandas as pd

import sqlite3
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)


# Create db connection.
cnx = sqlite3.connect('database.sqlite')



season = '2014/2015'
league_id = 21518


df = pd.read_sql_query("SELECT * FROM Match", cnx)

df = pd.read_sql_query("SELECT * FROM Match WHERE league_id = (?) AND season = (?)", cnx, params=(league_id,season,))


df.head(8)
```

```python
labels = df.apply(lambda x: res(x['home_team_goal'],x['away_team_goal']), axis =1 )
labels
```

```python
df_total_features.head()

def res(x,x1):
    if x > x1:
        return 'H'
    elif x < x1:
        return 'A'
    else:
        return 'D'









import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns
sns.countplot(labels,label="Count")

plt.show()





labels = labels[np.invert(df_total_features.isnull().any(axis=1)).values]

# drop rows with missing values

df_total_features.dropna(inplace=True)
# summarize the number of rows and columns in the dataset

# Create Training and Test Sets and Apply Scaling

from sklearn.model_selection import train_test_split

X = df_total_features
y = labels


random_seed=1

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=random_seed)



from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
scaler = MinMaxScaler()
scaler = StandardScaler()


X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)





# Building models



## Logistic regression

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))

print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))



from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)

print('Accuracy of SVM classifier on training set: {:.2f}'
     .format(svm.score(X_train, y_train)))
print('Accuracy of SVM classifier on test set: {:.2f}'
     .format(svm.score(X_test, y_test)))

## Let's try to convert the label to dummy variables


from sklearn.model_selection import train_test_split
y_A = pd.get_dummies(y)['A']
X_train, X_test, y_train_A, y_test_A = train_test_split(X, y_A,random_state=random_seed)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train_A)

print('Accuracy of Logistic regression classifier D on training set: {:.2f}'
     .format(logreg.score(X_train, y_train_A)))

print('Accuracy of Logistic regression classifier D on test set: {:.2f}'
     .format(logreg.score(X_test, y_test_A)))

from sklearn.model_selection import train_test_split
y_H = pd.get_dummies(y)['H']
X_train, X_test, y_train_H, y_test_H = train_test_split(X, y_H,random_state=random_seed)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train_H)

print('Accuracy of Logistic regression classifier H on training set: {:.2f}'
     .format(logreg.score(X_train, y_train_H)))

print('Accuracy of Logistic regression classifier H on test set: {:.2f}'
     .format(logreg.score(X_test, y_test_H)))

from sklearn.model_selection import train_test_split
y_D = pd.get_dummies(y)['D']
X_train, X_test, y_train_D, y_test_D = train_test_split(X, y_D,random_state=random_seed)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train_D)

print('Accuracy of Logistic regression classifier D on training set: {:.2f}'
     .format(logreg.score(X_train, y_train_D)))

print('Accuracy of Logistic regression classifier D on test set: {:.2f}'
     .format(logreg.score(X_test, y_test_D)))

```
