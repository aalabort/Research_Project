## Merging both datasets

```python
import pandas as pd

dataset_1 = 'df_testing_clustering_app5_3_[1729]_seasons_[\'15_16\'].csv'
dataset_2 = 'df_testing_clustering_app5_5_[1729]_seasons_[\'15_16\'].csv'
df_1 = pd.read_csv('data/'+ dataset_1)
df_2 = pd.read_csv('data/'+ dataset_2)



cols1 = df_1.columns
cols2 = df_2.columns


repeated_cols = []
for col1 in cols1:
    if col1 in cols2:
        repeated_cols.append(col1)


repeated_cols = set(repeated_cols)

labels = df_1['labels']
date = df_1['date']
repeated_cols.remove('uid')
repeated_cols.remove('cluster_team_home')
repeated_cols.remove('cluster_team_away')
print(repeated_cols)



df_1 = df_1.drop(columns=repeated_cols)
df_2 = df_2.drop(columns=repeated_cols)


```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_total_features=df_1.merge(df_2, left_on='uid', right_on='uid',how='left')


df_total_features['labels'] = labels
df_total_features['date'] = date
```

```python
df_total_features.head()
```

```python
location_to_file = 'data/'
df_total_features.to_csv(location_to_file+f'df_merge_{dataset_1[:10]}_{dataset_2[:10]}.csv',index=False)
```
